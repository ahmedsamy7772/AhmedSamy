```python
# ----------------- INSTALL DEPENDENCIES -----------------
!pip install ultralytics
!pip install mediapipe
!pip install opencv-python-headless

# ----------------- IMPORT LIBRARIES -----------------
import cv2
from ultralytics import YOLO
import mediapipe as mp
import numpy as np
from google.colab.patches import cv2_imshow
from IPython.display import display, Javascript, Image
from base64 import b64decode
from google.colab.output import eval_js # Ensure this is imported

# ----------------- FUNCTION TO CAPTURE PHOTO FROM BROWSER WEBCAM -----------------
def take_photo(filename='photo.jpg', quality=0.8):
    js = Javascript('''
        async function takePhoto(quality) {
          const div = document.createElement('div');
          const video = document.createElement('video');
          const button = document.createElement('button');
          button.textContent = 'Capture';
          div.appendChild(video);
          div.appendChild(button);
          document.body.appendChild(div);

          const stream = await navigator.mediaDevices.getUserMedia({video: true});
          video.srcObject = stream;
          await video.play();

          video.width = 640;
          video.height = 480;

          await new Promise((resolve) => button.onclick = resolve);

          const canvas = document.createElement('canvas');
          canvas.width = video.videoWidth;
          canvas.height = video.videoHeight;
          canvas.getContext('2d').drawImage(video, 0, 0);
          stream.getTracks().forEach(track => track.stop());
          div.remove();
          return canvas.toDataURL('image/jpeg', quality);
        }
    ''')
    display(js)
    data = eval_js('takePhoto({})'.format(quality))
    binary = b64decode(data.split(',')[1])
    with open(filename, 'wb') as f:
        f.write(binary)
    return filename

# ----------------- CONFIGURATION -----------------
model_path = "/content/runs/detect/cheating_detector222/weights/best.pt"  # path to your trained YOLO model

labels_dict = {
    "Ahmed20221154": "Ahmed20221154", # These should match your actual trained class names
    "Ramy20223509": "Ramy20223509"
}

# Extra objects to detect (make sure your model has these classes too)
extra_objects = ["mobile", "watch", "paper"]

# ----------------- LOAD MODEL -----------------
model = YOLO(model_path)

# ----------------- INIT MEDIA PIPE -----------------
mp_face = mp.solutions.face_mesh
face_mesh = mp_face.FaceMesh(static_image_mode=False, max_num_faces=1)

# ----------------- CAPTURE PHOTO -----------------
print("Please grant camera access and click 'Capture' to take a photo.")
photo_file = take_photo()
print("Photo captured!")

# Read the captured image
frame = cv2.imread(photo_file)
if frame is None:
    print("Failed to read captured image.")
else:
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # ----------------- YOLO PREDICTION -----------------
    results = model.predict(frame_rgb, conf=0.5, verbose=False)[0]

    detected_info = []
    for box, cls_id, score in zip(results.boxes.xyxy, results.boxes.cls, results.boxes.conf):
        cls_name = model.names[int(cls_id)]

        x1, y1, x2, y2 = map(int, box)
        color = (0,255,0) # Default green
        display_name = cls_name

        if cls_name in labels_dict:
            display_name = labels_dict[cls_name]
            detected_info.append(f"{display_name} with confidence {score:.2f}")
        elif cls_name in extra_objects:
            color = (0,0,255) # Red for extra objects / potential cheating
            cv2.putText(frame, f"{cls_name.upper()}! Cheating?", (x1, y2+20), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0,0,255), 2)
            detected_info.append(f"Potential cheating object: {cls_name} with confidence {score:.2f}")
        else:
            color = (0, 165, 255) # Orange for unknown person/object not in labels_dict
            detected_info.append(f"Unknown object/person: {cls_name} with confidence {score:.2f}")

        cv2.rectangle(frame, (x1,y1), (x2,y2), color, 2)
        cv2.putText(frame, f"{display_name} {score:.2f}", (x1,y1-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

    # ----------------- FACE MESH FOR TALKING / LOOKING AWAY -----------------
    face_results = face_mesh.process(frame_rgb)
    if face_results.multi_face_landmarks:
        landmarks = face_results.multi_face_landmarks[0].landmark
        # Simple logic: mouth open = talking (landmarks 13 upper lip, 14 lower lip)
        mouth_dist = np.linalg.norm(np.array([landmarks[13].x, landmarks[13].y]) - np.array([landmarks[14].x, landmarks[14].y]))
        if mouth_dist > 0.05:
            cv2.putText(frame, "Talking - might be cheating", (50,50), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,0,255),2)
            detected_info.append("Alert: User might be talking.")
        # Simple logic: nose X vs face center = looking away
        nose_x = landmarks[1].x
        if nose_x < 0.4 or nose_x > 0.6:
            cv2.putText(frame, "Looking away - might be cheating", (50,80), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,0,255),2)
            detected_info.append("Alert: User might be looking away.")

    # ----------------- DISPLAY FRAME IN COLAB -----------------
    cv2_imshow(frame)

    # ----------------- PRINT DETECTED INFORMATION -----------------
    if detected_info:
        print("\nDetected items and alerts:")
        for info in detected_info:
            print(f"- {info}")
    else:
        print("No specific individuals or cheating objects detected in the frame. person: Unkown")
```

```python
# ---------------- IMPORTS ----------------
from ultralytics import YOLO
import cv2
from IPython.display import display, Image, clear_output
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import output
from base64 import b64decode

# ---------------- CONFIG ----------------
model_path = "/content/runs/detect/cheating_detector222/weights/best.pt"
model = YOLO(model_path)

# Labels dictionary: persons in your model
labels_dict = {
    "Ahmed": "Ahmed20221154",
    "Ramy": "Ramy20223509"
}

# Extra objects for cheating alerts
extra_objects = ["mobile", "watch", "paper"]

# ---------------- FUNCTION TO GET LABEL ----------------
def get_label_name(label):
    return labels_dict.get(label, "Unknown")

# ---------------- CAPTURE FRAME FROM COLAB WEBCAM ----------------
from google.colab.output import eval_js

def capture_frame():
    js = """
    async function capture() {
      const video = document.createElement('video');
      const stream = await navigator.mediaDevices.getUserMedia({video: {width: 640, height: 480}});
      video.srcObject = stream;
      await video.play();

      const canvas = document.createElement('canvas');
      canvas.width = video.videoWidth;
      canvas.height = video.videoHeight;
      canvas.getContext('2d').drawImage(video, 0, 0);
      stream.getTracks().forEach(track => track.stop());
      return canvas.toDataURL('image/jpeg').substring(23);
    }
    capture();
    """
    data = eval_js(js)
    binary = b64decode(data)
    return cv2.imdecode(np.frombuffer(binary, np.uint8), cv2.IMREAD_COLOR)

# ---------------- RUN LOOP ----------------
print("Press Ctrl+C to stop the loop in Colab.")
try:
    while True:
        frame = capture_frame()
        if frame is None:
            print("No frame captured")
            continue

        # Predict with YOLO
        results = model.predict(frame, conf=0.5)[0]

        # Draw results
        for r in results.boxes:
            x1, y1, x2, y2 = map(int, r.xyxy[0])
            cls_id = int(r.cls[0])
            label = model.names[cls_id]

            # Default green box for known persons
            color = (0, 255, 0)
            display_text = get_label_name(label)

            # Extra objects 竊 red box + alert
            if label in extra_objects:
                color = (0, 0, 255)
                display_text = f"Cheating: {label}"

            # Draw box and label
            cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
            cv2.putText(frame, display_text, (x1, y1-10), cv2.FONT_HERSHEY_SIMPLEX, 0.7, color, 2)

        # Optional: simple talking / looking away alerts using mouth/nose landmarks
        # You can add MediaPipe FaceMesh here later if you want more advanced detection

        # Display in Colab
        clear_output(wait=True)
        cv2_imshow(frame)

except KeyboardInterrupt:
    print("Stopped.")
```
